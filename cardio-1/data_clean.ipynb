```python
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('Dataset/Cardio_Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df_cleaned = df.drop(['id', 'age','bp_category'], axis=1)
```

```python
df_cleaned.head()
```

```python
label_encoder = LabelEncoder()
df_cleaned['bp_category_encoded'] = label_encoder.fit_transform(df_cleaned['bp_category_encoded'])
```

```python
df_cleaned.head()
```

```python
df.isnull().sum()
```

```python
df_cleaned.to_csv(('data_cleaned_Cardio_Data.csv'), index=False)
```

```python
df_cleaned.describe()
```

Histograms for continuous variables to check the distribution.

```python
df_cleaned.hist(figsize=(16, 10), bins=20)
plt.show()
```

Box plots to identify outliers.

```python
df_cleaned.boxplot(figsize=(16, 8))
plt.show()
```

Bar plot to visualize frequency distribution of categorical columns.

```python
df_cleaned['bp_category_encoded'].value_counts().plot(kind='bar')
plt.show()
```

identifying relationships between numerical variables.

```python
correlation = df_cleaned.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.show()
```

box plots or Z-scores to detect outliers.

```python
z_scores = stats.zscore(df_cleaned.select_dtypes(include=['float64', 'int64']))
df_cleaned[(z_scores > 3).any(axis=1)] 
```
